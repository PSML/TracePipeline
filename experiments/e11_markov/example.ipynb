```python
def isSquare(P):
    return all (len (row) == len (P) for row in P)    

def isProb(P): 
    #Rows must sum to 1.
    for i in P.sum(axis=1):
        if not abs(i-1) < 10e-7:
            return False
    
    #Probabilities must be non-negative. 
    if not np.all((0 <= P)):
        return False
    return True
    
def is_markov(P):
    #A transition matrix for a markov process is square.
    if not isSquare(P): return False
    
    #It/'s rows all sum to one.
    if not isProb(P): return False

    return True
    
def markov_process(P, num_steps = 100):
    '''Returns a list generated by a markov process.'''
    assert is_markov(P)
    #Pick a random start state.
    state = random.randint(0, len(P)-1)
    trace = []
    for i in range(num_steps):
        trace += [state]
        #From [0, 1, ... #cols-1] pick according to xition matrix.
        state = np.random.choice(np.arange(0,P.shape[1]), p = P[state])
    return np.array(trace)
```

```python
P = np.array([
        [.1,  .3, .6],
        [.9, .05, .05],
        [.5,  .1, .4]
    ])
l = markov_process(P, num_steps=10000)
print(np.unique(l, return_counts=True)[1] / l.sum())

print(P.sum(axis=0) / P.sum())
```
